```python
#Criei uma função pra limpar as colunas pra evitar erros
def limpar_colunas(df):
    df.columns = df.columns.str.strip()
    return df

df_est = limpar_colunas(df_est)
df_cnae = limpar_colunas(df_cnae)
df_municipios = limpar_colunas(df_municipios)
```

```python
#Nova analise das colunas
print(df_est.columns)
print(df_cnae.columns)
print(df_municipios.columns)
```

```python
#Convertendo os dados das colunas
df_est["cnaeprincipal"] = pd.to_numeric(df_est["cnaeprincipal"], errors="coerce")
df_cnae["codigocnae"] = pd.to_numeric(df_cnae["codigocnae"], errors="coerce")
df_est["codigomunicipio"] = pd.to_numeric(df_est["codigomunicipio"], errors="coerce")
df_municipios["codigomunicipio"] = pd.to_numeric(df_municipios["codigomunicipio"], errors="coerce")

df_est["datainicioatividade"] = pd.to_datetime(
    df_est["datainicioatividade"], format="%Y%m%d", errors="coerce"
)
```

```python
print(df_est["datainicioatividade"].head())
```

```python
#Verificando os valores 
for col in ['cnaeprincipal', 'codigomunicipio', 'datainicioatividade']:
    print(f"Coluna: {col}")
    print(f" - Tipo: {df_est[col].dtype}")
    print(f" - Valores não convertidos (NaN/NaT): {df_est[col].isna().sum()}")
    print(f" - Alguns valores:\n{df_est[col].head()}")
    print("\n")
```

```python
#Join 
df = df_est.merge(
    df_cnae[["codigocnae", "cnae_descricao"]],
    left_on="cnaeprincipal",
    right_on="codigocnae",
    how="left"
)

df = df.merge(
    df_municipios[["codigomunicipio", "nomemunicipio"]],
    on="codigomunicipio",
    how="left"
)
```

```python
#Traduzir situação cadastral
map_situacao = {
    1: "NULA",
    2: "ATIVA",
    3: "SUSPENSA",
    4: "INAPTA",
    8: "BAIXADA"
}

df["situacao"] = df["situacaocadastral"].map(map_situacao)
df["situacao"].value_counts()
```

```python
#Filtrar setor de alimentação
palavras = [
    "restaurante", "lanchonete", "bar", "padaria",
    "aliment", "bebida", "pizzaria", "café", "buffet"
]

padrao = "|".join(palavras)

df_food = df[
    df["cnae_descricao"].fillna("").str.lower().str.contains(padrao)
]
```

```python
#Verificação de empresas ativas
ativas = df_food[df_food["situacaocadastral"] == 2]
print(len(ativas))
```

```python
#Crescimento mais ou menos presente 
segmentos = df_food["cnae_descricao"].value_counts()

print("Mais presentes:\n")
print(segmentos.head(10))

print("\nMenos presentes: \n")
print(segmentos.tail(10), "\n")
```

```python
#Crescimento ao longo do tempo
crescimento = (
    df_food["datainicioatividade"]
    .dt.year
    .value_counts()
    .sort_index()
)

print(crescimento)
```

```python
# Motivos de empresas baixadas/inaptas

problemas = df_food[
    df_food["situacaocadastral"].isin([4, 8])
]

motivos = (
    problemas["codigomotivosituacao"] 
    .value_counts()
)

print(motivos.head(10))
```
